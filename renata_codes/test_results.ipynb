```python
# Comparison between position arrays from test2, test3 and test4

for condition in ['dp','dn','mp','mn']:
    positions = [np.load(f'/media/renata/renata_data_2/Data_Behavior/data_notebook/tests/positions_{condition}_test{i}.npy', allow_pickle=True)
                for i in range(2,5)]
    
    assert len(positions[0])==len(positions[1])
    assert len(positions[1])==len(positions[2])

    num_traj = len(positions[0])

    for j in range(num_traj):
        print(np.all(positions[0][j] == positions[1][j]))
        print(np.all(positions[1][j] == positions[2][j]))
```
